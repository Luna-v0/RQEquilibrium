```python
import pyspiel
import pandas as pd
import numpy as np
```

```python
games = pyspiel.registered_games()
games
```

```python
df_games = [
{ "id":game.short_name,
  "name":game.long_name,
  "min_n_players": game.min_num_players,
  "max_n_players": game.max_num_players,
  "type": str(game.information).split(".")[1].lower(),
  "is_deterministic": str(game.chance_mode).split(".")[1].lower() == "deterministic"
} for game in games]


df = pd.DataFrame(df_games)
df.type = df.type.astype('string')
df
```
